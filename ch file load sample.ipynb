```python
import struct
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
len_offset = 248 # ヘッダ長
```

```python
data = open("/mnt/recordings/ExperimentData/2018-05-02/1Ecg.ch1", "rb").read()
```

```python
header_format = ""
for i in range( (len(data) - len_offset) // 2 ):
    header_format += "h"
```

```python
signal = np.array(struct.unpack_from(header_format, data, len_offset), dtype=np.float)
```

```python
plt.plot(signal[:2000])
```
